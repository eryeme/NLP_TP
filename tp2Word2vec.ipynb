```python
import gensim
from gensim.models import Word2Vec
from gensim.utils import simple_preprocess
import nltk
nltk.download('punkt')
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
path = '/content/text'
with open(path, 'r') as file:
    content = file.read()
content
```

```python
# Prétraitement du texte
processed_text = [simple_preprocess(sentence) for sentence in nltk.sent_tokenize(content)]
processed_text
```

```python
# Entraînement du modèle Word2Vec
modele = Word2Vec(sentences=processed_text, vector_size=100, window=5, min_count=1, workers=4)
```

```python
#Représentation vectorielle d'un mot
mot = 'civilizations'
vecteur = modele.wv[mot]
print(f"la représentation vectorielle du mot {mot} est: {vecteur}")
```

```python
#Similarité entre 2 mots
mot1 = 'civilizations'
mot2 = 'cultures'
similarite = modele.wv.similarity(mot1, mot2)
print(f"Similarité entre '{mot1}' et '{mot2}' : {similarite}")
```

```python
#mots contextuels (les plus similaires) pour un mot central donné
mot_central = 'marrakech'
mots_contextuels = modele.wv.most_similar(mot_central, topn=5) #la valeur de topn indique combien de mots similaires on veut
print(f"Mots contextuels pour le mot central '{mot_central}' : {mots_contextuels}")
```
